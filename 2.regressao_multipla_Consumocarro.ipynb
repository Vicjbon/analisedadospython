## Relacionando Consumo combustivel (mpg) com outras variáveis (cilindro ; cilindradas; horse power...)
## Regressão linear múltipla e Previsão

```python
# Importação das bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import statsmodels.formula.api as sm
```

```python
# Carregamento da base de dados e exclusão da primeiro coluna
base = pd.read_csv('mt_cars.csv')
base.shape
```

```python
#mpeg consumo, cyl clindros, disp cilindradas, hp
base.head()
```

```python
#exclui coluna dos carros 
base = base.drop(['Unnamed: 0'], axis = 1)
```

```python
# Criação de X e Y: variável independente e variável dependente
# Cálculo da correlação entre X (cilindradas) e Y (consumo)
X = base.iloc[:, 2].values #coluna disp - cilindradas
y = base.iloc[:, 0].values #coluna mpg - consumo
correlacao = np.corrcoef(X, y) #correlacionando as variaveis
correlacao
correlacao = np.corrcoef(X, y)[0, 1]
# Cria figura
plt.figure(figsize=(5,5))
# Gráfico de dispersão
plt.scatter(X, y, s=70, alpha=0.7, edgecolors='k')
# Linha de regressão vermelha
m, b = np.polyfit(X, y, 1)
plt.plot(X, m*X + b, color='red', linewidth=2)
# Eixos e título
plt.xlabel("Cilindradas (disp)")
plt.ylabel("Consumo (mpg)")
plt.title(f"Correlação = {correlacao:.2f}")
plt.grid(True)
plt.show()
```

```python
# Mudança do formato de X para o formato de matriz (necessário para versões mais recentes do sklearn)
X = X.reshape(-1, 1)
```

```python
# Criação do modelo, treinamento, visualização dos coeficientes e do score do modelo
modelo = LinearRegression()
modelo.fit(X, y)
```

```python
#Interceptação
modelo.intercept_
```

```python
#inclinação
modelo.coef_
```

```python
#score R^2 - O quanto o modelo é explicado pela variável explicativa X
modelo.score(X, y)
```

```python
# Geração das previsões de consumo
previsoes = modelo.predict(X)
previsoes
nova_cilindrada = np.array([[237]])  # precisa estar no formato 2D - Coloque quantas cilindradas X para prever o consumo Y
previsao = modelo.predict(nova_cilindrada)

print(f"Consumo previsto para 237 cilindradas: {previsao[0]:.2f} mpg")
```

```python
# Criação do modelo, utilizando a biblioteca statsmodel 
#podemos ver r ajustadodo r2
modelo_ajustado = sm.ols(formula = 'mpg ~ disp', data = base)
modelo_treinado = modelo_ajustado.fit()
modelo_treinado.summary()
```

```python
print(modelo_treinado.predict(pd.DataFrame({'disp': [217]}))[0])
```

```python
# Visualização dos resultados
plt.scatter(X, y)
plt.plot(X, previsoes, color = 'red')
```

```python
# Previsão para somente um valor
modelo.predict([[200]])
```

```python
# Criação de novas variáveis X1 e Y1 e novo modelo para comparação com o anterior - REGRESSÃO MULTILINEAR
# 3 variáveis dependentes para prever mpg: cyl	disp	hp
X1 = base.iloc[:, 1:4].values
X1
```

```python
y1 = base.iloc[:, 0].values
modelo2 = LinearRegression()
modelo2.fit(X1, y1)
#R^2
modelo2.score(X1, y1) # O quanto o modelo é explicado pelas variaveis X
```

```python
# Criação do modelo ajustado com mais atributos (regressão linear múltipla)
#usando stats models
modelo_ajustado2 = sm.ols(formula = 'mpg ~ cyl + disp + hp', data = base)
modelo_treinado2 = modelo_ajustado2.fit()
modelo_treinado2.summary()
```

```python
# Previsão de um novo registro - Calculando com base nessas variaveis X
novo = np.array([4, 200, 100])
novo = novo.reshape(1, -1)
modelo2.predict(novo)
```

```python
novo_exemplo = pd.DataFrame({
    'cyl': [6],
    'disp': [237],
    'hp': [100]
})

# Fazendo a previsão
previsao = modelo_treinado2.predict(novo_exemplo)
print(f"Consumo previsto: {previsao.iloc[0]:.2f} mpg")
```
